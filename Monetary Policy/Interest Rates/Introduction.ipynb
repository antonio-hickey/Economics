# Monetary Policy
- Interest Rates - A porportional payment on a loan, typically serviced anually.

### How Interest Rates are used as a tool to guide the economy:
 - Lowering rates stimulates the flow of credit, which in return grows the economy. Low rates create an environment of cheap money, but low interest rates hurt savers. We can see how lowering interest rates would hurt savers as they will now be payed a smaller porportion of their capital. So who does low interest rates help? Spenders, low rates make loans very accessible and cheap.
 
 - Raising rates drains the flow of credit, which in return shrinks the economy. Why would you want to raise interest rates? Countering inflationary pressure. When the Dollar is in danger of losing to much value, higher rates create an environment of expensive money. Expensive money hurts spenders and rewards savers. Higher rates incite saving by paying bigger porportions of capital.

### The 10 Year Treasury Note
- A Government bond, where you loan money to the government expecting to be paid back in full plus interest in 10 years. Treasury Bonds are considered the safest heaven asset due to the good faith in the U.S Government and it being very unlikely they defualt on the debt. Investors see this as a safe spot to park money.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
from datetime import datetime
import numpy as np
from matplotlib.ticker import PercentFormatter

data = pd.read_csv('DGS10.csv')
data.head()

x_ = data['Date'].values
x_ = [datetime.strptime(x, '%Y-%m-%d') for x in x_]

y = data['Rate'].values.astype(float)

years = mdates.YearLocator()   # every year
months = mdates.MonthLocator()  # every month
years_fmt = mdates.DateFormatter('%Y')

fig, ax = plt.subplots()
ax.plot(x_, y, color='teal', lw=3.5)

# format the ticks
ax.xaxis.set_major_locator(years)
ax.xaxis.set_major_formatter(years_fmt)
ax.xaxis.set_minor_locator(months)
ax.yaxis.set_major_formatter(PercentFormatter())


# round to nearest years.
#datemin = np.datetime64(data['Date'][100], 'Y')
#datemax = np.datetime64(data['Date'][-1:], 'Y') + np.timedelta64(1, 'Y')
#ax.set_xlim(datemin, datemax)

locator = mdates.YearLocator()
plt.gca().xaxis.set_major_locator(locator)
plt.gcf().autofmt_xdate()
plt.title('U.S 10 Year Treasury Note')
plt.ylabel('Interest Rate')
plt.show()
```

### Notes from this chart:
- We can see obvious economic growth and a strong economy from 2016 until the end of 2018. infering this from interest rates rising, but as we see money became to expensive in 2019 threatning the economy, so we see a drop in rates to try and stimulate growth and inflation.

- Interest rates are near zero, in fact real rates (inflation adjusted) are negative. What does this mean? Right now investors buying 10 Year T-Note's expect to take a small loss, but overall keeping their portfolio relatively safe. Best way to think of this is like an insurance payment.

### Recap of Interest Rate basics:
- Lower rates = stimultion, spending, inflationary

- Higher rates = draining, saving, deflationary

# Negative Interest Rates
- Negative Interest Rate Policy (NIRP) - Where the borrower is payed interest to borrow money, while the creditor is paying interest to loan their money out. This is a radical use of interest rates as a last resort to stimulate growth, spending,  and inflation. NIRP are currently being used by Bank of Japan (BoJ), and the Swiss National Bank (SNB), but this radical use of monetary policy has not seemed to work as they have failed to create inflation in their respected economies.

### Swiss National Bank:

```python
data2 = pd.read_csv('DGS11.csv')

x_1 = data2['Date'].values
x_1 = [datetime.strptime(x, '%Y-%m-%d') for x in x_1]

Rates = data2['Rates'].values

Fx = data2['Fx'].values

Cpi = data2['Cpi'].values

years = mdates.YearLocator()   # every year
months = mdates.MonthLocator()  # every month
years_fmt = mdates.DateFormatter('%Y')

fig, ax = plt.subplots()
ax.plot(x_1, Rates, color='teal', lw=3.5)
ax.set_ylabel('Interest Rate')
ax.axhline(y=0, color='black', linestyle='--')

ax2 = ax.twinx()
color = 'firebrick'
ax2.set_ylabel('Swiss Franc vs U.S Dollar', color=color)  # we already handled the x-label with ax1
ax2.plot(x_1, Fx, color=color, lw=3.5)
ax2.tick_params(axis='y', labelcolor=color)


# format the ticks
ax.xaxis.set_major_locator(years)
ax.xaxis.set_major_formatter(years_fmt)
ax.xaxis.set_minor_locator(months)
ax.yaxis.set_major_formatter(PercentFormatter())


# round to nearest years.
#datemin = np.datetime64(data['Date'][100], 'Y')
#datemax = np.datetime64(data['Date'][-1:], 'Y') + np.timedelta64(1, 'Y')
#ax.set_xlim(datemin, datemax)

locator = mdates.YearLocator()
plt.gca().xaxis.set_major_locator(locator)
plt.gcf().autofmt_xdate()
plt.title('Swiss National Bank 10 Year Bond vs Swiss Franc')
plt.show()
```

- Notice as rates go negative the swiss franc becomes very flat, but actually deflates the Franc as rates go negative. Here negative interest rates did not seem to operate as intended to create inflation. 

### The big question:
### How far negative can interest rates go?

The Swiss rates have gone the lowest in the world with a low of -0.975%, the problem was the Franc was getting even stronger the further negative rates went. I don't see how central banks would go lower than -1.00% as Switzerland and Japans expiraments with negative rates have not seemed to work as intended. Negative rates may create a monetary black hole of sorts, when interest rates go negative people have to pay interest to hold cash in their bank account, As you'd expect people withdraw their deposit's with banks to avoid having to service interest payments. This is called a run on the bank's which then pushes bank's into insolvency.
